```python
### Sreelakshmi - Assignment 2  - Sales Dataset

# PART 1 - Data Pre-processing #

import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. Load the data set
salesDataUpload = pd.read_csv(r'C:\Users\jkast\OneDrive\Documents\Python\Sales_Data_Set\archive\train.csv')
df = pd.DataFrame(salesDataUpload)
df.info()

# 2. Data Cleaning
df=df.fillna(0)
print(df.head(20))
```

```python
# PART 2 - Univariate Analysis #

# 1. Distribution Analysis

# Histogram for continuous variable:
sns.histplot(data=df, x='Purchase', bins =20, kde=True)

```

```python
# Bar charts for categorical variables:
sns.barplot(data=df, x='Age', y='Purchase', hue='Gender', order=df['Age'])
plt.show()
sns.barplot(data=df, x='Stay_In_Current_City_Years', y='Purchase', hue='City_Category')
plt.show()
sns.barplot(data=df, x='Stay_In_Current_City_Years', y='Purchase', hue='Gender')
plt.show()
sns.barplot(data=df, x='Age', y='Purchase', hue='Stay_In_Current_City_Years')
plt.show()
sns.barplot(data=df, x='City_Category', y='Purchase', hue='Stay_In_Current_City_Years')
plt.show()

#sns.barplot(data=df, x='Product_Category_1', y='Purchase')
#plt.show()
```
